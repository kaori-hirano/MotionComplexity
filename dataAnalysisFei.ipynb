## Frame Work

```python
# removes parenthesis
def extract_and_convert(s):
    numbers = re.findall(r"[-+]?\d*\.\d+|\d+", s)
    return float(numbers[0]) if numbers else None

# gets cols 
columns_to_convert = ['currentTime','headx','heady','headz','headyaw','headpitch','headroll','hmdx','hmdy','hmdz','hmdyaw','hmdpitch','hmdroll','joystick1','joystick2','hmd_linear1','hmd_linear2','hmd_linear3','hmd_angular1','hmd_angular2','hmd_angular3','restrict1','restrict2','restrict3','restrict4','frameRate']

def frame_processing(dfs):
    dfs_copy = {}
    for key, df in dfs.items():
        # make copy of df
        df_copy = df.copy()

        for col in columns_to_convert:
            df_copy[col] = df_copy[col] = df_copy[col].astype(str).apply(extract_and_convert)

        dfs_copy[key] = df_copy

    return dfs_copy

proframes = frame_processing(frame_dfs)
```

```python
def combine_dfs(combined_event, frame_dfs): 
    # makes new df for the included information 
    final_df = pd.DataFrame()
      
    # gets every row (which is a trial) in event 
    for index, row in combined_event.iterrows(): 
        # stores trial info  
        start = row['StartTime']
        end = row['EndTime']
        key = row['participantId']

        # saves row information to be appended to all frames 
        copied_event_row = pd.DataFrame({
                'participantId': [row['participantId']],
                'TrialId': [row['TrialId']],
                'FMSScore': [row['FMSScore']],
                'Condition': [row['Condition']]
        })

        # gets frame for that row 
        frame_df = frame_dfs[key].copy()

        # gets right text file information about broad vs narrow
        fov_value = get_bn(key)
        fov_data = pd.DataFrame({'fov': [fov_value]})

        # row[['participantId', 'TrialId', 'FMSScore', 'Condition']]
        
        # get frames within range
        frames_in_trial = frame_df.loc[frame_df['currentTime'].between(start, end)].reset_index(drop=True)
        
        # gets right num of event 
        num_rows = len(frames_in_trial)
        repeated_event_row = pd.concat([copied_event_row] * num_rows, ignore_index=True)

        # right num of fov condition 
        repeated_fov = pd.concat([fov_data] * num_rows, ignore_index=True)

        row_data = pd.concat([repeated_event_row, repeated_fov, frames_in_trial], axis=1)
        
        # add row to final df 
        final_df = pd.concat([final_df, row_data], ignore_index=True)
        
        # print(row_data.head(2))

    return final_df
```
